## News

##### Get latest news
/news
https://api.tradingeconomics.com/news?c=guest:guest

```python
te.getNews(output_type='df')
```

##### Get news by country
/news/country/{countries}
https://api.tradingeconomics.com/news/country/united%20states?c=guest:guest

```python
te.getNews(country='united states',output_type='df')
```

##### Get news by indicator
/news/indicator/{indicators}
https://api.tradingeconomics.com/news/indicator/inflation%20rate?c=guest:guest

```python
te.getNews(indicator='inflation rate',output_type='df')
```

##### Get news by country and indicator
/news/country/{countries}/{indicators}
https://api.tradingeconomics.com/news/country/united%20states/inflation%20rate?c=guest:guest

```python
te.getNews(country='united states', indicator='inflation rate', output_type='df')
```

##### Paginate news list by specifying start index and list size
/news?c=guest:guest&limit={list_size}&start={start_index}
http://api.tradingeconomics.com/news?c=guest:guest&limit=15&start=10

```python
te.getNews(start='15', limit='15', output_type='df')
```

##### By start and end date

```python
te.getNews(start_date='2021-02-02', end_date='2021-02-05', output_type='df')
```

##### by start and end date and indicator

```python
te.getNews(indicator='inflation rate', start_date='2021-02-02', end_date='2021-03-03',output_type='df')
```

##### By multiple indicators

```python
te.getNews(indicator=['inflation rate', 'gdp'], output_type='df')
```

##### By multiple countries and multiple indicators

```python
te.getNews(country=['brazil','canada'], indicator=['Housing Starts', 'Stock Market'], start_date='2021-02-02', end_date='2021-03-03', output_type='df')
```
