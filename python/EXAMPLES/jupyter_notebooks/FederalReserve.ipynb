# Federal Reserve

##### 1 List of all US states
/fred/states
https://api.tradingeconomics.com/fred/states?c=guest:guest

```python
te.getFedRStates(output_type='df')
```

##### 2  List of all counties per state
/fred/counties/{state}
https://api.tradingeconomics.com/fred/counties/arkansas?c=guest:guest

```python
te.getFedRStates(county = 'arkansas', output_type = 'df')
```

##### 3 - Data by Symbol
/fred/snapshot/symbol/{symbol}
https://api.tradingeconomics.com/fred/snapshot/symbol/ALLMARGATTN?c=guest:guest

```python
te.getFedRSnaps(symbol='ALLMARGATTN', output_type='df')
```

##### 4 data by URL. Example: "url=/united-states/all-marginally-attached-workers-for-tennessee-fed-data.html"

https://api.tradingeconomics.com/fred/snapshot/url?c=guest:guest&url=/united-states/all-marginally-attached-workers-for-tennessee-fed-data.html

```python
te.getFedRSnaps(url='url=/united-states/all-marginally-attached-workers-for-tennessee-fed-data.html', output_type='df')
```
